```python
time_normal_count = 0
time_total_count = 0

for results_file in time_series_results:
    df = pd.read_csv(base_directory + "time_series/" + results_file)

    # Get the 100 results
    results = df[time_results_column].values

    time_normal_count += metrics.shapiro_p(results)
    time_total_count += 1

time_non_normal_count = time_total_count - time_normal_count
```

```python
total_non_normal_count = image_non_normal_count + time_non_normal_count
total_count = image_total_count + time_total_count

image_b_result = binomtest(
    k=image_non_normal_count, n=image_total_count, p=p_baseline, alternative="greater"
)

time_b_result = binomtest(
    k=time_non_normal_count, n=time_total_count, p=p_baseline, alternative="greater"
)

total_b_result = binomtest(
    k=total_non_normal_count, n=total_count, p=p_baseline, alternative="greater"
)

print("Image Classification Non-Normal Count:", image_non_normal_count)
print("Image Classification Total Count:", image_total_count)
print(
    "Image Classification Binomial Test Result:",
    metrics.safe_round(image_b_result.pvalue, round_digits),
)
print("===============")
print("Time Series Non-Normal Count:", time_non_normal_count)
print("Time Series Total Count:", time_total_count)
print(
    "Time Series Binomial Test Result:",
    metrics.safe_round(time_b_result.pvalue, round_digits),
)
print("===============")
print("Total Non-Normal Count:", total_non_normal_count)
print("Total Count:", total_count)
print(
    "Total Binomial Test Result:",
    metrics.safe_round(total_b_result.pvalue, round_digits),
)
```
