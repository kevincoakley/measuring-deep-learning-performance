```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "../scopus/regularization/regularization.csv"
df = pd.read_csv(file_path)

# Remove the "All Methods" row
df_filtered = df[df["Regularization Method"] != "All Methods"]

# Set the categories and values for the bar chart
categories = df_filtered["Regularization Method"]
#no_filter = df_filtered["No Filter"]
image = df_filtered["Image"]
time_series = df_filtered["Time Series"]

# Create the bar chart with wider bars for better readability
plt.figure(figsize=(14, 6))
bar_width = 0.4  # Increase bar width for better visibility
spacing = 0.6  # Keep sufficient spacing between groups
x = range(len(categories))

#bars1 = plt.bar([i * (3 * bar_width + spacing) for i in x], no_filter, width=bar_width, label="No Filter")
bars2 = plt.bar([i * (3 * bar_width + spacing) + bar_width for i in x], image, width=bar_width, label="Image")
bars3 = plt.bar([i * (3 * bar_width + spacing) + 2 * bar_width for i in x], time_series, width=bar_width, label="Time Series")

# Add value labels above each bar
#for bars in [bars1, bars2, bars3]:
for bars in [bars2, bars3]:
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom', fontsize=12)

# Set labels and title
plt.xlabel("Methods", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.title("Scopus Results for the Number of Published Papers on Regularization Methods", fontsize=18)
plt.xticks([i * (3 * bar_width + spacing) + bar_width for i in x], categories, rotation=45, ha="right")
plt.legend(fontsize=14)
plt.tick_params(labelsize=14)

# Save the plot
plt.savefig("plots/scopus/regularization.pdf", dpi=600, bbox_inches='tight')  # Save the plot with proper bounding
```
